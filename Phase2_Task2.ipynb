```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as mtp
```

```python
# Read data from an train HTML file
dfs = pd.read_html("https://docs.google.com/spreadsheets/u/1/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml", header=0)
# Assuming there is only one table in the HTML, you can access it using index 0
df1 = dfs[0]
# Save the data to a CSV file
df1.to_csv('train.csv', index=False, header=False)

# Read data from an test HTML file
dfss = pd.read_html("https://docs.google.com/spreadsheets/u/1/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pubhtml", header=0)
# Assuming there is only one table in the HTML, you can access it using index 0
df2 = dfss[0]
# Save the data to a CSV file
df2.to_csv('test.csv', index=False, header=False)
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
print(train_df)
print(test_df)
```

```python
#For train dataset
# Drop the first column and null values
trimmed_trained_data = train_df.drop(train_df.columns[0], axis=1)
trimmed_trained_data = trimmed_trained_data.dropna()
# Print the trimmed DataFrame
print(trimmed_trained_data)

#For test dataset
# Drop the first column and null values
trimmed_test_data = test_df.drop(test_df.columns[0], axis=1)
trimmed_test_data = trimmed_test_data.dropna()
# Print the trimmed DataFrame
print(trimmed_test_data)
```

```python
#Accessing the train dataset columns
x1= trimmed_trained_data.iloc[:, :-1].values
y1= trimmed_trained_data.iloc[:, 1].values

#Accessing the test dataset columns
x2= trimmed_test_data.iloc[:, :-1].values
y2= trimmed_test_data.iloc[:, 1].values
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x1, y1)
```

```python
#Prediction of Test and Training set result
y_pred= regressor.predict(x2)
x_pred= regressor.predict(x1)
```

```python
#visualizing the Train set results
mtp.scatter(x1, y1, color="green")
mtp.plot(x1, x_pred, color="red")
mtp.title("X VS Y (Training Dataset)")
mtp.xlabel("X")
mtp.ylabel("Y")
mtp.show()
```

```python
#visualizing the Test set results
mtp.scatter(x2, y2, color="blue")
mtp.plot(x1, x_pred, color="red")
mtp.title("X vs Y (Test Dataset)")
mtp.xlabel("X")
mtp.ylabel("Y")
mtp.show()
```
