```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('/content/sample_data/dataset - netflix1 (2).csv')

#Count plot of show types
plt.figure(figsize=(10,6))
sns.countplot(x='type', data=dataset, palette='viridis')
plt.title('Count of Show Types on Netflix')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.show()
```

```python
#Bar plot of top 10 countries with most content on Netflix
top_countries = dataset['country'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
#Bar plot of top 10 directors with most content on Netflix
top_directors = dataset['director'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
top_directors.plot(kind='bar', color='salmon')
plt.title('Top 10 Directors with Most Content on Netflix')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
#Count plot of content added over the years
dataset['date_added'] = pd.to_datetime(dataset['date_added'])
dataset['year_added'] = dataset['date_added'].dt.year
plt.figure(figsize=(12,6))
sns.countplot(x='year_added', data=dataset, palette='pastel')
plt.title('Content Added Over the Years on Netflix')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.show()
```

```python
#Count plot of content ratings
plt.figure(figsize=(10,6))
sns.countplot(x='rating', data=dataset, palette='rocket')
plt.title('Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Function to preprocess the duration column
def preprocess_duration(duration):
    if 'Season' in duration:
        return int(duration.split(' ')[0])
    elif 'min' in duration:
        return int(duration.replace(' min', ''))
    else:
        return None

# Preprocess the duration column
dataset['duration'] = dataset['duration'].apply(preprocess_duration)

#Histogram of content durations
plt.figure(figsize=(12,6))
sns.histplot(dataset['duration'].dropna(), kde=True, color='lightblue')
plt.title('Distribution of Content Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```
